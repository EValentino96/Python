###Safe Input

```
# making sure that the game doesn't break after the wrong input

def secureInput(nums = (), isFloat = False, string = ''):
    
    if string != '': # if there's something to print out print it out
        Print(string)

    s.write(color.b)
    ans = input('-> ')
    s.write(color.e)
    
    if isFloat:
        try:
            ans = float(ans)  # if it's meant to be a float make it a float
        except:
            return secureInput(isFloat = True, string = 'Looks like you didn\'t enter a number, please try again:')

        if ans < 0 or ans > 1: # make sure its between the numbers its meant to be
            return secureInput(isFloat = True, string = 'Looks like you didn\'t enter a decimal between 0 and 1, please try again:')
        
        return ans

    elif not is_empty(nums): # if there's stuff in nums do this
        try:
            ans = int(ans) # try to turn it into a number
        except:
            return secureInput(nums=nums, string = 'Looks like you didn\'t enter a number, please try again:')

        if not exists(ans,nums): # make sure it's in the range of numbers it must be
            return secureInput(nums = nums, string = 'Looks like your number doesn\'t match any of the options, please try again:')

        return ans
    
    else: # otherwise was expecting a yes or no answer

        ans = ans.lower() # making things easier, set to lower case

        if len(ans) > 0: # as long as it's not an empty string...

            if not exists(ans[0], ('y','n')): # check if it's a yes answer or a no answer (only check first letter)
                return secureInput(string='Looks like you didn\'t answer yes or no, please try again:')

        else: # if it's an empty string then they must have hit enter.
            return secureInput(string='Looks like you hit enter instead of answering, please try again:')
        return ans[0]
```

```
def useItem():
    global player

    print()
    # check if there are even any items to use & RETURN
    if is_empty(player.items):
        Print(color.bla + 'You have no items to use at this time.')
        return

    # if player has something to use:
    else:
        # display all the items the player has, plus an extra if they change their mind
        Print('Which item would you like to use?')

        print(color.b + color.w + '[ ',end='')
        for i in range(len(player.items)):
            print('('+str(i+1)+') '+player.items[i].name.ljust(10,)+'\t',end='')
            if i > 1 and np.mod(i,5) == 0:
                print('\n  ',end='')
        i += 1
        print('\t('+str(i+1)+') NEVERMIND ]'+color.e) # extra option that takes you out of using items

    index = secureInput(nums = tuple(range(1, i+2)))-1

    if index >= len(player.items): # if player doesn't want to use any items, exit useItem menu & RETURN
        return
    
    item = player.items[index-1] # otherwise, find the item the player chose

    # before each item, let the player know the stats of the item.
    
    # if the item is normal food type add to player hp, up to max no more:
    if item.kind == 'f':
        print(color.b + 'Would you like to eat '+ color.g + item.name.upper() + color.e + color.b +' and gain' + color.g + ' +' + str(item.buff) + ' HP' + color.e + color.b + '? [y/n]' + color.e)
        food = secureInput()

        if food.lower() == 'n': # if they don't want to use THIS item, take them back one step to the item menu
            useItem()

        # if this item increases above health reset to max otherwise add to player hp
        if player.hitpoints + item.buff >= player.maxHitpoints: 
            player.hitpoints = player.maxHitpoints
        else:
            player.hitpoints = player.hitpoints + item.buff
        print(color.b + '\n--YOU have eaten ' + color.g + item.ame.upper() + color.e + color.b +' and gained' + color.g +' +' + str(item.buff) + ' HP' + color.e + color.b + '--\n' + color.e)
        
        # remove item from inventory
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)
        return
    
    # if the item is normal drink type add to player ap, up to max no more
    elif item.kind == 'd':
        print(color.b + 'Would you like to drink '+ color.c + item.name.upper() + color.e + color.b +' and gain' + color.c + ' +' + str(item.buff) + ' AP' + color.e + color.b + '?[y/n]\n' + color.e)
        drink = secureInput()

        if drink.lower() == 'n': # if they don't want to use THIS item, take them back one step to the item menu
            useItem()
        
        # if this item increases above health reset to max otherwise add to player ap
        if player.attackpoints + item.buff >= player.maxAttackpoints:  
            player.attackpoints = player.maxAttackpoints
        else:
            player.attackpoints = player.attackpoints + item.buff
        
        print(color.b + '\n--YOU have drank ' + color.c + item.name.upper() + color.e + color.b +' and gained' + color.c +' +' + str(item.buff) + ' AP' + color.e + color.b + '--\n' + color.e)
        
        # remove item from inventory
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)
        return
    
    # if item is special food, increase player max hp, and give full hp
    elif item.kind == 'sf':
        print(color.b + 'Would you like to eat '+ color.G + item.name.upper() + color.e + color.b +' and gain' + color.G + ' +' + str(item.buff) + ' HP INCREASE' + color.e + color.b + '?[y/n]\n' + color.e)
        food = secureInput()

        if food.lower() == 'n': # if they don't want to use THIS item, take them back one step to the item menu
            useItem()
        
        player.maxHitpoints = player.maxHitpoints + item.buff
        player.hitpoints = player.maxHitpoints
        
        print(color.b + '\n--YOU have eaten ' + color.G + item.name.upper() + color.e + color.b +' and gained' + color.G +' +' + str(item.buff) + ' HP INCREASE' + color.e + color.b + '--\n' + color.e)
        
        # remove item from inventory
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)        
        return
        
    # if item is special drink, increase player max ap, and give full ap
    elif item.kind == 'sd':
        print(color.b + 'Would you like to drink '+ color.C + item.name.upper() + color.e + color.b +' and gain' + color.C + ' +' + str(item.buff) + ' AP INCREASE' + color.e + color.b + '?[y/n]\n' + color.e)
        drink = secureInput()

        if drink.lower() == 'n':
            useItem()
        
        player.maxAttackpoints = player.maxAttackpoints + item.buff
        player.attackpoints = player.maxattackpoints
        
        print(color.b + '\n--YOU have drank ' + color.C + item.name.upper() + color.e + color.b +' and gained' + color.C +' +' + str(item.buff) + ' AP INCREASE' + color.e + color.b + '--\n' + color.e)
        
        # remove item from inventory
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)        
        return
    
    # if item is an object: if it can be applied to attacks, apply otherwise don't use
    elif item.kind == 'o':

        if item.buff == 0: # if the item isn't meant for an attack
            print('This item cannot be applied to any attack.')
            useItem()
        
        print(color.b + 'Would you like to apply ' + color.r + item.name + color.e + color.b + '? (' + 
              color.r + item.name + color.e + color.b + ' will change damage by: ' + color.r + str(item.buff[0]) + color.e + color.b + ' and cost by: ' + color.r + str(item.buff[1]) + color.e +
              color.b + ')\n' + '[ (1) ' + player.attack1.name + '    (2) ' + player.attack2.name + '    (3) ' + player.attack3.name + '    (4) none ]\n' + color.e)
        
        # have player pick which attack to apply item to, have option for neither which takes player back to useItem menu
        attack = secureInput(nums = (1,2,3,4))

        if attack == 4:
            useItem()

        # apply to each attack and let player know the stats of the attack
        if attack == 1:
            player.attack1.dmg = 0 if player.attack1.dmg + item.buff[0] < 0 else player.attack1.dmg + item.buff[0]
            player.attack1.cost = 0 if player.attack1.cost + item.buff[1] < 0 else player.attack1.cost + item.buff[1]
            print(color.b + '--YOU have applied ' + color.r + item.name.upper() + color.e + color.b + ' to your ' + color.r 
              + player.attack1.name + color.e + color.b + ' ATTACK, modified buffs are: ' + color.r + str(player.attack1.dmg) + color.e + color.b + 
              ' Damage, ' + color.r + str(player.attack1.cost) + color.e + color.b + ' Cost--\n' + color.e)
        
        elif attack == 2:
            player.attack2.dmg = 0 if player.attack2.dmg + item.buff[0] < 0 else player.attack2.dmg + item.buff[0]
            player.attack2.cost = 0 if player.attack2.cost + item.buff[1] < 0 else player.attack2.cost + item.buff[1]
            print(color.b + '--YOU have applied ' + color.r + item.name.upper() + color.e + color.b + ' to your ' + color.r 
              + player.attack2.name + color.e + color.b + ' ATTACK, modified buffs are: ' + color.r + str(player.attack2.dmg) + color.e + color.b + 
              ' Damage, ' + color.r + str(player.attack2.cost) + color.e + color.b + ' Cost--\n' + color.e)
        
        elif attack == 3:
            player.attack3.dmg = 0 if player.attack3.dmg + item.buff[0] < 0 else player.attack3.dmg + item.buff[0]
            player.attack3.cost = 0 if player.attack3.cost + item.buff[1] < 0 else player.attack3.cost + item.buff[1]
            print(color.b + '--YOU have applied ' + color.r + item.name.upper() + color.e + color.b + ' to your ' + color.r 
              + player.attack3.name + color.e + color.b + ' ATTACK, modified buffs are: ' + color.r + str(player.attack3.dmg) + color.e + color.b + 
              ' Damage, ' + color.r + str(player.attack3.cost) + color.e + color.b + ' Cost--\n' + color.e)
        
        # remove item from inventory
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)     
        return
```

###Buy Items

```
def buyItems(shopInventory, special=False):
    global player
    
    print()
    Print(' SHOP '.center(55,'-'))
    
    print(color.b + color.w)
    for i in range(len(shopInventory)):
        
        if player.money < (shopInventory[i].buff + abs(shopInventory[i].effect)):
            print(color.bla + '(' + str(i+1) + ') ' + shopInventory[i].name.ljust(20,'-'), end='')
            print('Price: '.rjust(30,'-') + str(shopInventory[i].buff + abs(shopInventory[i].effect)) + color.w)
        else:
            print('(' + str(i+1) + ') ' + shopInventory[i].name, end='')
            print('Price: ' + str(shopInventory[i].buff + abs(shopInventory[i].effect)))
  
    i += 1
    print('(' + str(i+1) + ') NONE' + color.e)

    print()
    Print('YOUR MONEY: ' + str(player.money))

    print()
    Print('Which item would you like to check?')

    itemdex = secureInput(nums = tuple(range(1, i+2))) - 1
    
    if itemdex >= len(shopInventory):
        return

    item = shopInventory[itemdex]
    cost = item.buff + abs(item.effect)

    kind = ''
    if item.kind == 'f':
        kind = 'food'
    if item.kind == 'd':
        kind = 'drink'
    if item.kind == 'sf':
        kind = 'special food'
    if item.kind == 'sd':
        kind = 'special drink'
    if item.kind == 'o':
        kind = 'object'

    print()
    Print('Item: ' + item.name)
    Print('Type: ' + kind)
    Print('Player Boost: '+ str(item.buff))
    Print('Affinity Boost: '+ str(item.effect))

    if player.money < cost:
        print()
        Print('You don\'t have enough money to buy this.')
        return buyItems(shopInventory, special)

    print()
    Print('Would you like to buy ' + color.P + item.name + color.w + ' for ' + color.G + str(cost) + color.w + ' money? [y/n]')

    ans = secureInput()

    if ans == 'n':
        return buyItems(shopInventory, special)
    
    else:
        if player.money < cost:
            print()
            Print('You don\'t have enough money to buy this.')
            return buyItems(shopInventory, special)
        
        player.money = player.money - cost
        player.items = player.items + (item,)
        print()
        Print('You have just bought an item! Thank you for your service. Would you like to buy something else? [y/n]')

        again = secureInput()

        if again:
            return buyItems(shopInventory, special)
        else:
            return        
```

```
def battle(opponent, playerTurn, mainFight = False):
    global player

    mn = -1; mx = 2; rnd = int(mn + (mx - mn)*random())
    wait = 1

    # YOU LOSE
    if player.hitpoints <= 0:
        sleep(wait)
        print()
        Print('--You have been defeated in battle--')
        if opponent.fought == 0:
            opponent.fought = 1 # you lost so record that as a win for opponent

        # reset player stats
        player.hitpoints = player.maxHitpoints
        player.actionpoints = player.maxActionpoints
        opponent.hitpoints = opponent.maxHitpoints

        if mainFight: # if it was a main fight, go to custom dialogue
            return
        else:
            return returnToCheckpoint() # else go back to checkpoint


    # YOU WIN
    elif opponent.hitpoints <= 0:
        # refresh opponent HP
        opponent.hitpoints = opponent.maxHitpoints
    
        sleep(wait)
        print()
        Print('--' + opponent.name + ' has been defeated!--')
        if opponent.fought == 0:
            opponent.fought = -1 # record as a loss for opponent

        sleep(wait)
        print()
        Print('--YOU WIN!--')

        sleep(wait)
        print()
        Print(color. g + '** GAIN: +5 MONEY **')
        player.hitpoints = player.maxHitpoints
        player.actionpoints = player.maxActionpoints
        player.money = player.money + 5
        
        if mainFight: # if it was a main fight, go to custom dialogue
            return
        else:
            return returnToCheckpoint() # else go back to checkpoint

    # DISPLAY WHOSE TURN IT IS
    turn = ' ' + opponent.name + '\'s TURN ' 
    if playerTurn:
        turn = ' YOUR TURN '
    print()
    print(color.b + color.w + turn.center(35,'-') + color.e)

    # DISPLAY STATS BEFORE DOING SOMETHING
    print()
    print(color.b + color.blu + color. u + 'YOU' + color.e + color.b + color.w + '\t\tvs\t' + color.r + color.u + opponent.name + color.e + color.b + color.blu + '\n' + 
          'HP: ' + str(player.hitpoints) + '/' + str(player.maxHitpoints) + '\t\t' + color.r + 'HP: ' + str(opponent.hitpoints) + '/' + str(opponent.maxHitpoints) + '\n' + 
          color.blu + 'AP: ' + str(player.actionpoints) + '/' + str(player.maxActionpoints) + '\t\t' + color.r + 'AFFINITY: ' + str(opponent.affinity) + '\n' + 
          color.blu + 'ITEMS: ' + str(len(player.items)) + '\t\t' + color.r + 'FAVORS: ' + opponent.favors + color.e)
    
    #PLAYER GOES
    if playerTurn:
        
        repeat = PlayerTurn(opponent) #continue battle or end battle (because player chose to leave)

        if repeat == False: # if player chose to leave
            return
        else:
           return battle(opponent, False, mainFight)

    # OPPONENT GOES
    else:
        OpponentTurn(opponent)
        return battle(opponent, True, mainFight)
    
```

```
def OpponentTurn(opponent):
    # The big idea is: the more the opponent hates you, the stronger attacks they use and the less they hit
    hit = False
    if random() < f(opponent.affinity/4):
        hit = True

    wait = 1
    sleep(wait) 
                    ## ALL ATTACKS HAVE SAME DYNAMIC

    # if they like you too much, they will use the weak attack only------ AFF > 10
    if opponent.affinity > 10:
        print()
        print(color.b + '--' + opponent.name + ' uses '+ color.r + opponent.attack1[0] + color.e + color.b + ' ATTACK on YOU--'+color.e)
            
        if not hit: # attack misses
            sleep(wait)
            print(color.b + '--YOU dodge ' + color.r + opponent.attack1[0] + color.e + color.b + ' ATTACK--' + color.e)
            return 

        sleep(wait)
        print()    
        Print(opponent.attack1[2]) # they words for that attack 

        sleep(wait)
        print()
        print(color.b + '--' + opponent.name + ' has inflicted '+ color.r + str(opponent.attack1[1]) +color.e + color.b + ' damage on YOU--'+color.e)
        player.hitpoints = player.hitpoints - opponent.attack1[1]

    # if sort of like you, they use 1 or 2------------------------------- 0 <= AFF <= 10
    elif opponent.affinity >= 0 and opponent.affinity <= 10:

        attack = opponent.attack2   
        if random() < 0.5:
            attack = opponent.attack1

        print()
        print(color.b + '--' + opponent.name + ' uses '+ color.r + attack[0] + color.e + color.b + ' ATTACK on YOU--'+color.e)
        
        if not hit:
            sleep(wait)
            print(color.b + '--YOU dodge the ATTACK--' + color.e)
            return 

        sleep(wait)
        print()
        Print(attack[2])
        
        sleep(wait)
        print()
        print(color.b + '--' + opponent.name + ' has inflicted '+ color.r + str(attack[1]) +color.e + color.b + ' damage on YOU--'+color.e)
        player.hitpoints = player.hitpoints - attack[1]
        
    # if sort of hates you, they use 2 or 3------------------------------- -10 <= AFF <= 0
    elif opponent.affinity >= -10 and opponent.affinity <= 0:

        attack = opponent.attack3    
        if random() < 0.5:
            attack = opponent.attack2

        print()
        print(color.b + '--' + opponent.name + ' uses '+ color.r + attack[0] + color.e + color.b + ' ATTACK on YOU--'+color.e)
            
        if not hit:
            sleep(wait)
            print(color.b + '--YOU dodge the ATTACK--' + color.e)
            return
        
        sleep(wait)
        print()
        Print(attack[2])
        
        sleep(wait)
        print()
        print(color.b + '--' + opponent.name + ' has inflicted '+ color.r + str(attack[1]) +color.e + color.b + ' damage on YOU--'+color.e)
        player.hitpoints = player.hitpoints - attack[1]
        
    # if they hate you too much, they will only use attack 3-------------- AFF < -10
    else: 
        print()
        print(color.b + '--' + opponent.name + ' uses '+ color.r + opponent.attack3[0] + color.e + color.b + ' ATTACK on YOU--'+color.e)
            
        if not hit:
            sleep(wait)
            print(color.b + '--YOU dodge the ATTACK--' + color.e)
            return 
        
        sleep(wait)
        print()
        Print(opponent.attack3[2])
        
        sleep(wait)
        print()
        print(color.b + '--' + opponent.name + ' has inflicted '+ color.r + str(opponent.attack3[1]) +color.e + color.b + ' damage on YOU--'+color.e)
        player.hitpoints = player.hitpoints - opponent.attack3[1]
    return
```

####Player Turn

```
def PlayerTurn(opponent): # returns a 'repeat' boolean if it's okay to continue battle (only false when player tries to leave)
    global player
    global tutorial

    # does player even have enough ap to attack?
    haveAP = False if player.actionpoints < player.attack1.cost else True

    if tutorial: # if this is set to true, give the tutorial talk
        print()
        print(color.b + color.w +'[ ',end='')
    
        if not haveAP:
            print(color.bla + '(1) FIGHT' + color.w, end='')
        else:
            print('(1) FIGHT',end='')
    
        print('    ',end='')

        if is_empty(player.items) and not haveAP:
            print(color.bla + '(2) GIVE' + color.w, end='')
        else:
            print('(2) GIVE',end='')

        print('    (3) OTHER ]' + color.e)

    
        Print(color.b + color.r + 'BUFF: Here you are given 4 options, you can read them. Fighting is pretty obvious, shows your attacks, their damage and how much AP each costs'+
              ' if at any point you don\'t have enough AP to fight, FIGHT will be '+color.bla + 'GREY'+color.r+'.',fp=True)
        print()
        Print(color.b + color.r + 'BUFF: Actually this moves me to my next point, GIVE is if you are a weakling and want to GIVE your opponent either an item or a compliment and only giving compliments'+
              ' costs 2 AP, however if you don\'t have any items to give or don\'t have 2 AP, GIVE will be '+color.bla + 'GREY' + color.r + '.',fp=True)
        
        print()
        Print(color.b + color.r + 'BUFF: Now finally, OTHER is as it implies other options that aren\'t fighting or giving. There you will have the option to use any item on yourself, leave the battle,'+
              ' or meditate. ITEMS can either be used for your HP, AP, total HP, total AP or your attacks. LEAVING is where you ATTEMPT to leave like a WHIMP, but if the opponent'+
              ' really hates you, they won\'t let you get away. Also leaving will not count as a loss or a win, so if the opponent is someone you MUST fight then you will have to do the fight over'+
              ' MEDITATING allows you to be one with your inner troubles and returns to you some AP.',fp=True)
        
        print()
        Print(color.b + color.r + 'BUFF: That\'s all there is to it, DOT, oh and don\'t worry about picking the wrong option, you can go back from any menu!', fp=True)

        tutorial = False # once done with tutorial set to false

    # give the player the options of FIGHT GIVE OTHER
    print()
    print(color.b + color.w +'[ ',end='')
    
    if not haveAP:
        print(color.bla + '(1) FIGHT' + color.w, end='')
    else:
        print('(1) FIGHT',end='')
    
    print('    ',end='')

    if is_empty(player.items) and not haveAP:
        print(color.bla + '(2) GIVE' + color.w, end='')
    else:
        print('(2) GIVE',end='')

    print('    (3) OTHER ]' + color.e)

    pick = secureInput(nums = (1, 2, 3)) # have player pick what they want to do

    if pick == 1: # ----------------------------------------- YOU PICK FIGHT --------------------------

        repeat = Fight(opponent, haveAP) 

        if repeat:# if player accidentally picked fight OR didn't have enough AP for any attack bring them back here
            return PlayerTurn(opponent)
        
    elif pick == 2: # ------------------------------------- YOU PICK GIVE ---------------------------------------

        repeat = Give(opponent)
        if repeat: # if player accidentally picked give
            return PlayerTurn(opponent)

    elif pick == 3: # ----------------------------------- YOU PICK OTHER ------------------------------------
        return Other(opponent) # returns true OR false when player wants to leave

    return True
```

####Fight

```
def Fight(opponent, haveAP): # returns 'repeat' boolean, true for repeat player turn, false for go to opponents turn
    global player
    mn = 1; mx = 6; r = int(mn + (mx - mn)*random())

    print()
    if not haveAP: # you don't have enough action points for any attack go back to player turn
            Print('You do not have enough AP to use an attack!')
            return True

    # random change to miss (1 in 5)
    hit = False if r == 1 else True
    
    #pick the attack to use or nevermind
    print(color.b + 'CHOOSE YOUR ATTACK:\n' + color.r + 
        '[ (1) ' + player.attack1.name + ' ATTACK | Damage ' + str(player.attack1.damage) + ' | Cost ' + str(player.attack1.cost) + '\n' + 
        '  (2) ' + player.attack2.name + ' ATTACK | Damage ' + str(player.attack2.damage) + ' | Cost ' + str(player.attack2.cost) + '\n' +
        '  (3) ' + player.attack3.name + ' ATTACK | Damage ' + str(player.attack3.damage) + ' | Cost ' + str(player.attack3.cost) + '\n' +
        '  (4) NEVERMIND ]'+ color.e)
        
    atk = secureInput(nums = (1, 2, 3, 4))
    
    if atk == 4:
        return True # repeat turn (AKA go back to player menu)

    if atk == 1: # ---------------------------- YOU PICKED ATTACK 1
        print()
        # no need to check ap vs cost because that would be part of haveAP

        print(color.b + '--YOU use ' + color.r + player.attack1.name + color.e + color.b + ' ATTACK on ' + opponent.name + '--' + color.e)
                
        if not hit: # attack missed so go back and play opponent turn
            print(color.b + '--' + opponent.name + ' dodges your ATTACK--' + color.e)
            return False 
                
        print(color.b + '--YOU have inflicted '+ color.r + str(player.attack1.damage) + color.e + color.b + ' damange on ' + opponent.name + '--'+color.e)
        opponent.hitpoints = opponent.hitpoints - player.attack1.damage
        player.actionpoints = player.actionpoints - player.attack1.cost # no need to worry about going under because that was checked for at the start
            
    elif atk == 2: # -------------------------- YOU PICKED ATTACK 2                     ALL ATTACKS WORK THE SAME
        print()
        # make sure that the player has enough action points for this attack, otherwise go back to fighting menu
        if player.actionpoints < player.attack2.cost:
            Print('You do not have enough AP for this attack..!')
            return Fight(opponent, True)    

        print(color.b + '--YOU use ' + color.r + player.attack2.name + color.e + color.b + ' ATTACK on ' + opponent.name + '--' + color.e)

        if not hit:# attack missed so go back and play opponent turn
            print(color.b + '--' + opponent.name + ' dodges your ATTACK--' + color.e)
            return False
                
        print(color.b + '--YOU have inflicted '+ color.r + str(player.attack2.damage) + color.e + color.b + ' damange on ' + opponent.name + '--'+color.e)
        opponent.hitpoints = opponent.hitpoints - player.attack2.damage
        player.actionpoints = player.actionpoints - player.attack2.cost 

    # same as  attack 2
    elif atk == 3: # -------------------------- YOU PICKED ATTACK 3
        print()
        if player.actionpoints < player.attack3.cost:
            Print('You do not have enough AP for this attack..!')
            return Fight(opponent, True)

        print(color.b + '--YOU use ' + color.r + player.attack3.name + color.e + color.b + ' ATTACK on ' + opponent.name + '--' + color.e)

        if not hit:
            print(color.b + '--' + opponent.name + ' dodges your ATTACK--' + color.e)
            return False
                
        print(color.b + '--YOU have inflicted '+ color.r + str(player.attack3.damage) + color.e + color.b + ' damange on ' + opponent.name + '--'+color.e)
        opponent.hitpoints = opponent.hitpoints - player.attack3.damage
        player.actionpoints = player.actionpoints - player.attack3.cost

    return False
```

####Give

```
def Give(opponent): #returns 'repeat' boolean for repeating player turn if they accidentally picked give
    global player

    cost = 2 # cost for giving a compliment

    # print item black out if you don't have items (just to make player aware they can't give item)   
    print()
    Print('Choose what to give ' + color.r + '(giving compliment will cost ' + str(cost) + ' AP)' +color.w + ':')

    print(color.b + color.p + '[ ',end='')
    if is_empty(player.items):
        print(color.bla + '(1) ITEM' + color.p, end = '')
    else:
        print('(1) ITEM',end='')
    print('    ',end = '')
    if player.actionpoints < cost:
        print(color.bla + '(2) COMPLIMENT' + color.p, end='')
    else:
        print('(2) COMPLIMENT',end = '')
    print('    (3) NEVERMIND ]' + color.e)
    
    do = secureInput(nums = (1, 2, 3))

    if do == 3: # returns to player turn and repeats (AKA goes back to player menu)
        return True

    # randomness to if opponent will like what they are given
    likes = True if random() < f(opponent.affinity) else False
    
    if do == 1: # ---------------------- YOU GIVE ITEM
                
        if is_empty(player.items):  # if there's nothing to give go back to giving menu
            print()  
            Print(color.bla+'You have no items to give at this time.')
            return Give(opponent)
        
        print()
        Print('Which item would you like to give?')
        
        print(color.b + color.p + '[ ', end = '')
        for i in range(len(player.items)):
            print('('+str(i+1)+') '+player.items[i].name+'\t',end='')
        i += 1 # i = len(player.items)
        print('\t('+str(i+1)+') NEVERMIND ]'+ color.e) #the extra option
                
        index = secureInput(nums = tuple(range(1, i+2))) - 1
                
        if index >= len(player.items): # if player chose to give item by accident, goes back to give menu
            return Give(opponent)
                    
        item = player.items[index-1]

        print()    
        Print(color.p + '--YOU give '+ opponent.name + ' ' + item.name + '--') # give item to oppoent

        # remove item from player item list  
        itemlist = list(player.items)
        itemlist.remove(item)
        player.items = tuple(itemlist)
        
        # if its an item that the opponent favors, apply the appropriate effect
        if exists(item, opponent.favoriteItems):
            print()
            print(color.b + color.g + '--' + opponent.name + ' LOVES this GIFT--\n'+
            '--AFFINITY INCREASED A LOT--\n' +
            '--' + opponent.name + ' is WEAKER now--' + color.e)
                        
            opponent.affinity = opponent.affinity + item.effect
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] - item.effect, opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] - item.effect, opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] - item.effect, opponent.attack2[2])
        
        else: # otherwise leave it to random chance
            Liking(opponent, likes, 'GIFT')
                    
    elif do == 2:# ------------------------ YOU GIVE COMPLIMENT

        print()
        if player.actionpoints < cost: # check if player has enough AP to give a compliment
            Print('You do not have enough action points to give a compliment!')
            return Give(opponent)
        
        Print(color.p + '--YOU give ' + opponent.name + ' a COMPLIMENT--')
 
        Liking(opponent, likes, 'COMPLIMENT')  # leave it to random chance if opponent likes      
        player.actionpoints = player.actionpoints - cost

    return False # return false for not repeating player turn
```

```
def Liking(opponent, likes, xs):

    # if the opponent likes you a lot, their attack decreases by how much affinity they have with you
    if opponent.affinity > 5:
        print(color.b + color.g + '--' + opponent.name + ' likes the '+ xs +'--\n'+
            '--AFFINITY INCREASED--\n' +
            '--' + opponent.name + ' is WEAKER now--' + color.e)
                        
        opponent.affinity = opponent.affinity + 1
        opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] - opponent.affinity, opponent.attack1[2])
        opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] - opponent.affinity, opponent.attack2[2])
        opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] - opponent.affinity, opponent.attack2[2])

    # if the opponent likes you, there's a chance that they might not like the gift             
    elif opponent.affinity >= 0 and opponent.affinity <= 5:
        if likes:
            print(color.b + color.g + '--' + opponent.name + ' likes the '+ xs +'--\n'+
                '--AFFINITY INCREASED--\n' +
                '--' + opponent.name + ' is a little WEAKER now--' + color.e)
                            
            opponent.affinity = opponent.affinity + 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] - int(opponent.affinity/2), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] - int(opponent.affinity/2), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] - int(opponent.affinity/2), opponent.attack2[2])
        else:
            print(color.b + color.r + '--' + opponent.name + ' dislikes the '+ xs +'--\n'+
                '--AFFINITY DECREASED--\n' +
                '--' + opponent.name + ' is little STRONGER now--' + color.e)
                            
            opponent.affinity = opponent.affinity - 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] + int(opponent.affinity/3), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] + int(opponent.affinity/3), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] + int(opponent.affinity/3), opponent.attack2[2])

    # the rest mirror the top two                
    elif opponent.affinity >= -5 and opponent.affinity <= 0:
        if likes:
            print(color.b + color.g + '--' + opponent.name + ' likes the '+ xs +'--\n'+
                '--AFFINITY INCREASED--\n' +
                '--' + opponent.name + ' is somwhat WEAKER now--' + color.e)
                            
            opponent.affinity = opponent.affinity + 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] - abs(int(opponent.affinity/3)), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] - abs(int(opponent.affinity/3)), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] - abs(int(opponent.affinity/3)), opponent.attack2[2])
        else:
            print(color.b + color.r + '--' + opponent.name + ' dislikes the '+ xs +'--\n'+
                '--AFFINITY DECREASED--\n' +
                '--' + opponent.name + ' is somewhat STRONGER now--' + color.e)
                            
            opponent.affinity = opponent.affinity - 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] + abs(int(opponent.affinity/2)), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] + abs(int(opponent.affinity/2)), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] + abs(int(opponent.affinity/2)), opponent.attack2[2])
    else:
        if likes:
            print(color.b + color.g + '--' + opponent.name + ' likes the '+ xs +'--\n'+
                '--AFFINITY INCREASED--\n' +
                '--' + opponent.name + ' is slightly WEAKER now--' + color.e)
                    
            opponent.affinity = opponent.affinity + 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] - abs(int(opponent.affinity/4)), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] - abs(int(opponent.affinity/4)), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] - abs(int(opponent.affinity/4)), opponent.attack2[2])
        else:
            print(color.b + color.r + '--' + opponent.name + ' dislikes the '+ xs +'--\n'+
                '--AFFINITY DECREASED--\n' +
                '--' + opponent.name + ' is STRONGER now--' + color.e)
                        
            opponent.affinity = opponent.affinity - 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] + abs(opponent.affinity), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] + abs(opponent.affinity), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] + abs(opponent.affinity), opponent.attack2[2])
```

####Other

```
def Other(opponent): # returns true if battle should repeat (aka continue) else returns false
    global player

    print()
    Print('Choose your action: (Meditating increases your AP)')
    print(color.b + color.y + '[ (1) USE ITEM    (2) LEAVE    (3) MEDITATE    (4) NEVERMIND ]' + color.e)
    pik = secureInput(nums = (1,2,3,4))
    
    if pik == 4: # accidentally picked other
        return PlayerTurn(opponent)

    if pik == 1: # ----------------- YOU USE ITEM
        useItem() # go to use item menu
            
    elif pik == 2: # ---------------- YOU WALK AWAY
            
        # you attempt to leave battle
        print()
        Print(color.y + '--YOU try to leave BATTLE--')
        chase = True if random() < f(opponent.affinity) else False

        # if the opponent chanses you, you go back to battle    
        if chase:
            print(color.b + color.r + '--' + opponent.name + ' has chased YOU--\n'+
                '--AFFINITY DECREASED--\n' +
                '--' + opponent.name + ' is STRONGER now--' + color.e)
            opponent.affinity = opponent.affinity - 1
            opponent.attack1 = (opponent.attack1[0], opponent.attack1[1] + abs(int(opponent.affinity)), opponent.attack1[2])
            opponent.attack2 = (opponent.attack2[0], opponent.attack2[1] + abs(int(opponent.affinity)), opponent.attack2[2])
            opponent.attack3 = (opponent.attack3[0], opponent.attack3[1] + abs(int(opponent.affinity)), opponent.attack3[2])
            return True
            
        # otherwise, you successfully left battle
        Print(color.y + '--YOU have SUCCESSFULLY left the BATTLE--')
        # battle terminates but doesnt count as fully fighting opponent.
        # everyone's stats are refreshed (except player)
        opponent.hitpoints = opponent.maxHitpoints
        return False

    elif pik == 3: # ---------------- YOU MEDITATE
        print()

        if player.actionpoints == player.maxActionpoints: # if you already have max AP don't do anything
            Print(color.y + '--YOU already have max AP--')
        
        elif player.actionpoints + player.attack1.cost > player.maxActionpoints: # if you'll go over just set to max
            Print(color.y + '--YOU have medicated and are at FULL AP--')
            player.actionpoints = player.maxActionpoints
        
        else:                                                                      # otherwise add normally
            Print(color.y + '--YOU have meditated and increased your AP (+' + str(player.attack1.cost) + ')--')
            player.actionpoints = player.actionpoints + player.attack1.cost
    
    return True # return true to repeat battle
```

###Printing

```
def Print(xs, newline=True, fp=False):
    global speed

    words = xs.split()
    thresh = 5*20 # avg word length is 5, want approx 20 words per line
    wordlen = 0

    if fp: # this option is to create a 'typed' effect, originally called this fancy print hence why fpa
        for word in words:
            wordlen += len(word)

            if wordlen > thresh:
                print() # start a new line after the threshold
                wordlen = len(word) #the current words length restarts count

            for letters in word:
                print(letters, end = ''); sleep(speed) #print the letters slowly

            print(' ', end='') # put a space between words

    else:
        s.write(color.b + color.w) # make standard non typed text bold and white

        for word in words:
            wordlen += len(word)
                
            if wordlen > thresh:
                print() # start a new line after the threshold
                wordlen = len(word) # the current words length restarts count

            print(word + ' ', end='') # put a space between words
    
    s.write(color.e) # clear all color from typed and non typed

    # enter after this string? if yes do so else don't
    if newline:
        print()
    else:
        print(end='')
```
